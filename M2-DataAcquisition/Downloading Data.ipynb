```python
import requests

# URL of the dataset (Iris dataset from UCI Machine Learning Repository)
url = "https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data"

# Send a GET request to download the data
response = requests.get(url)

# Check if the request was successful
if response.status_code == 200:
    with open("iris.csv", "wb") as file:
        file.write(response.content)
    print("Data downloaded successfully!")
else:
    print(f"Failed to download data. Status code: {response.status_code}")
```

```python
from urllib.request import urlretrieve

# URL of the dataset
url = "https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data"

# Download the file
urlretrieve(url, "iris_urllib.csv")
print("Data downloaded successfully!")
```

```python
import requests

# API endpoint and API key (sign up at https://openweathermap.org/api to get your own API key)
api_key = "your_api_key_here"  # Replace with your API key
city = "Pittsburgh"
url = f"http://api.openweathermap.org/data/2.5/weather?q={city}&appid={api_key}"

# Send a GET request to the API
response = requests.get(url)

# Check if the request was successful
if response.status_code == 200:
    data = response.json()  # Assuming the API returns JSON data
    print("Data downloaded successfully!")
    print(data)  # Display the downloaded data
else:
    print(f"Failed to download data. Status code: {response.status_code}")
```

### HOW TO READ AN API ###
 
Go to the specification tab, those are the endpoints. Each endpoint can have various parameters it accepts

```python
import requests

headers = {'User-Agent' : 'myapp'}
# to see the alerts for Pennsylvania, use:
endpoint = 'https://api.weather.gov/alerts/active?area=PA'
# to see the zones for Pennsylvania, use:
endpoint = 'https://api.weather.gov/zones?area=PA'
# to see the stations for Pennsylvania, use:
endpoint = 'https://api.weather.gov/stations?state=PA'


response = requests.get(endpoint, headers = headers)
data = response.json()

# `features` contains the data we want.
print(data['features']) 

# show the first alert, formatted nicely
import json
print(json.dumps(data['features'][0], indent=2))
```
