## Handling Missing Data in Weather Sensor Data

Let's create a synthetic weather dataset that mimics real sensor data with missing values, then explore different methods to handle the missing data.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from datetime import datetime, timedelta

# Set random seed for reproducibility
np.random.seed(42)

# Create synthetic weather sensor data
n_days = 100
start_date = datetime(2024, 1, 1)
dates = [start_date + timedelta(days=i) for i in range(n_days)]

# Generate base weather data
temperature = np.random.normal(15, 10, n_days)  # Temperature in Celsius
humidity = np.random.normal(60, 20, n_days)     # Humidity percentage
pressure = np.random.normal(1013, 15, n_days)   # Atmospheric pressure in hPa
wind_speed = np.random.exponential(5, n_days)   # Wind speed in km/h
weather_conditions = np.random.choice(['Sunny', 'Cloudy', 'Rainy', 'Foggy'], n_days, 
                                    p=[0.4, 0.3, 0.2, 0.1])

# Create DataFrame
weather_data = pd.DataFrame({
    'date': dates,
    'temperature': temperature,
    'humidity': humidity,
    'pressure': pressure,
    'wind_speed': wind_speed,
    'condition': weather_conditions
})

# Introduce missing values to simulate sensor failures
# More aggressive missing data patterns

# Random missing values (sensor malfunction) - increased frequency
missing_temp = np.random.choice(n_days, size=int(n_days * 0.18), replace=False)
weather_data.loc[missing_temp, 'temperature'] = np.nan

missing_humidity = np.random.choice(n_days, size=int(n_days * 0.15), replace=False)
weather_data.loc[missing_humidity, 'humidity'] = np.nan

missing_pressure = np.random.choice(n_days, size=int(n_days * 0.12), replace=False)
weather_data.loc[missing_pressure, 'pressure'] = np.nan

missing_wind = np.random.choice(n_days, size=int(n_days * 0.20), replace=False)
weather_data.loc[missing_wind, 'wind_speed'] = np.nan

# Sequential missing values (sensor offline for periods)
weather_data.loc[15:22, 'humidity'] = np.nan     # 8-day humidity sensor failure
weather_data.loc[35:41, 'pressure'] = np.nan     # 7-day pressure sensor failure  
weather_data.loc[60:67, 'temperature'] = np.nan  # 8-day temperature sensor failure
weather_data.loc[80:84, 'wind_speed'] = np.nan   # 5-day wind sensor failure

# Multiple sensor failures (power outage scenarios)
weather_data.loc[25:28, ['temperature', 'pressure']] = np.nan
weather_data.loc[55:57, ['humidity', 'wind_speed']] = np.nan
weather_data.loc[72:74, ['temperature', 'humidity', 'pressure']] = np.nan

# Burst failures (intermittent connectivity issues)
burst_days1 = [10, 12, 14, 16]  # Every other day failure pattern
burst_days2 = [45, 47, 49, 51, 53]  # Another burst pattern
weather_data.loc[burst_days1, 'wind_speed'] = np.nan
weather_data.loc[burst_days2, 'humidity'] = np.nan

# Missing categorical data - increased
missing_condition = np.random.choice(n_days, size=int(n_days * 0.12), replace=False)
weather_data.loc[missing_condition, 'condition'] = None

# Additional weekend maintenance patterns (systematic missing data)
weekend_maintenance = [6, 7, 13, 14, 20, 21, 27, 28, 34, 35]  # Simulate weekend maintenance
weather_data.loc[weekend_maintenance, 'pressure'] = np.nan

print("Original Dataset with Missing Values:")
print(weather_data.head(15))
print(f"\nDataset shape: {weather_data.shape}")
print(f"\nMissing values per column:")
print(weather_data.isnull().sum())
print(f"\nPercentage of missing values:")
print((weather_data.isnull().sum() / len(weather_data) * 100).round(2))
```

### Method 1: Dropping Rows with Missing Values

```python
# Method 1: Drop rows with ANY missing values
method1_dropna_any = weather_data.dropna()
print("Method 1a - Drop rows with ANY missing values:")
print(f"Original shape: {weather_data.shape}")
print(f"After dropping: {method1_dropna_any.shape}")
print(f"Data loss: {len(weather_data) - len(method1_dropna_any)} rows ({((len(weather_data) - len(method1_dropna_any))/len(weather_data)*100):.1f}%)")

# Method 1b: Drop rows with ALL missing values (less aggressive)
method1_dropna_all = weather_data.dropna(how='all')
print(f"\nMethod 1b - Drop rows with ALL missing values:")
print(f"After dropping: {method1_dropna_all.shape}")
print(f"Data loss: {len(weather_data) - len(method1_dropna_all)} rows")

# Method 1c: Drop rows based on specific columns
method1_dropna_subset = weather_data.dropna(subset=['temperature', 'pressure'])
print(f"\nMethod 1c - Drop rows missing temperature OR pressure:")
print(f"After dropping: {method1_dropna_subset.shape}")
print(f"Data loss: {len(weather_data) - len(method1_dropna_subset)} rows ({((len(weather_data) - len(method1_dropna_subset))/len(weather_data)*100):.1f}%)")
```

### Method 3: Statistical Imputation (Mean, Median, Mode)

```python
# Method 3: Statistical Imputation
method3_mean = weather_data.copy()
method3_median = weather_data.copy()

# Fill numeric columns with mean
numeric_columns = ['temperature', 'humidity', 'pressure', 'wind_speed']
for col in numeric_columns:
    method3_mean[col] = method3_mean[col].fillna(method3_mean[col].mean())
    method3_median[col] = method3_median[col].fillna(method3_median[col].median())

# Fill categorical column with mode
mode_condition = weather_data['condition'].mode()[0]  # Get the most frequent value
method3_mean['condition'] = method3_mean['condition'].fillna(mode_condition)
method3_median['condition'] = method3_median['condition'].fillna(mode_condition)

print("SUMMARY STATISTICS COMPARISON - Statistical Imputation Methods")
print("=" * 70)

# Create comprehensive comparison of summary statistics for numeric columns
for col in numeric_columns:
    print(f"\n{col.upper()}:")
    print("-" * 40)
    
    # Calculate statistics for original data (excluding NaN)
    original_stats = weather_data[col].describe()
    mean_stats = method3_mean[col].describe()
    median_stats = method3_median[col].describe()
    
    # Create comparison DataFrame
    stats_comparison = pd.DataFrame({
        'Original': original_stats,
        'Mean_Imputed': mean_stats,
        'Median_Imputed': median_stats
    })
    
    print(stats_comparison.round(3))
    
    # Show the differences from original
    print(f"\nImputation values used:")
    print(f"Mean: {weather_data[col].mean():.3f}")
    print(f"Median: {weather_data[col].median():.3f}")
    
    print(f"\nDifference from original mean:")
    print(f"Mean Imputation: {(mean_stats['mean'] - original_stats['mean']):.3f}")
    print(f"Median Imputation: {(median_stats['mean'] - original_stats['mean']):.3f}")
    
    print(f"\nDifference from original std:")
    print(f"Mean Imputation: {(mean_stats['std'] - original_stats['std']):.3f}")
    print(f"Median Imputation: {(median_stats['std'] - original_stats['std']):.3f}")

print(f"\n" + "=" * 70)
print("CATEGORICAL VARIABLE:")
print(f"Condition mode used for imputation: {mode_condition}")
print(f"Missing values after imputation:")
print(f"Mean method: {method3_mean.isnull().sum().sum()}")
print(f"Median method: {method3_median.isnull().sum().sum()}")

print(f"\n" + "=" * 70)
print("KEY INSIGHTS:")
print("• Mean imputation preserves the original mean exactly")
print("• Median imputation may slightly shift the mean toward the median") 
print("• Both methods reduce variance (standard deviation)")
print("• Mean imputation more sensitive to outliers than median")
print("• Statistical imputation ignores temporal patterns in time series")
```

```python
# Method 4: Interpolation
method4_linear = weather_data.copy()
method4_polynomial = weather_data.copy()

# Linear interpolation for numeric columns
for col in numeric_columns:
    method4_linear[col] = method4_linear[col].interpolate(method='linear')
    
# Polynomial interpolation for numeric columns  
for col in numeric_columns:
    method4_polynomial[col] = method4_polynomial[col].interpolate(method='polynomial', order=2)

# For categorical data, use forward fill as interpolation doesn't work
method4_linear['condition'] = method4_linear['condition'].fillna(method='ffill')
method4_polynomial['condition'] = method4_polynomial['condition'].fillna(method='ffill')

print("SUMMARY STATISTICS COMPARISON - Interpolation Methods")
print("=" * 70)

# Create comparison of summary statistics for numeric columns
for col in numeric_columns:
    print(f"\n{col.upper()}:")
    print("-" * 40)
    
    # Calculate statistics for original data (excluding NaN)
    original_stats = weather_data[col].describe()
    linear_stats = method4_linear[col].describe()
    polynomial_stats = method4_polynomial[col].describe()
    
    # Create comparison DataFrame
    stats_comparison = pd.DataFrame({
        'Original': original_stats,
        'Linear_Interp': linear_stats,
        'Polynomial_Interp': polynomial_stats
    })
    
    print(stats_comparison.round(3))
    
    # Show the differences from original
    print(f"\nDifference from original mean:")
    print(f"Linear Interpolation: {(linear_stats['mean'] - original_stats['mean']):.3f}")
    print(f"Polynomial Interpolation: {(polynomial_stats['mean'] - original_stats['mean']):.3f}")
    
    print(f"\nDifference from original std:")
    print(f"Linear Interpolation: {(linear_stats['std'] - original_stats['std']):.3f}")
    print(f"Polynomial Interpolation: {(polynomial_stats['std'] - original_stats['std']):.3f}")

print(f"\n" + "=" * 70)
print("Missing values after interpolation:")
print(f"Linear method: {method4_linear.isnull().sum().sum()}")
print(f"Polynomial method: {method4_polynomial.isnull().sum().sum()}")

# Show the difference between interpolation methods
print(f"\n" + "=" * 70)
print("INTERPOLATION COMPARISON - Sample values (humidity, rows 18-28):")
interpolation_comparison = pd.DataFrame({
    'Original': weather_data.loc[18:28, 'humidity'],
    'Linear': method4_linear.loc[18:28, 'humidity'],
    'Polynomial': method4_polynomial.loc[18:28, 'humidity'],
    'Mean_Fill': method3_mean.loc[18:28, 'humidity']
})
print(interpolation_comparison.round(2))

print(f"\n" + "=" * 70)
print("KEY INSIGHTS:")
print("• Linear interpolation creates smooth transitions between known values")
print("• Polynomial interpolation can create more complex curves but may overshoot")
print("• Both methods preserve temporal relationships in time series data")
print("• Interpolation works best with ordered data and short gaps")
print("• May create false precision for long missing data sequences")
```

### Summary of Missing Data Handling Methods

**Method 1: Dropping Data**
- **Pros**: Simple, no assumptions about missing values, preserves data integrity for remaining observations
- **Cons**: Can lead to significant data loss, may introduce bias if missingness is not random, reduces statistical power
- **Best for**: When missing data is truly random and you have plenty of data

**Method 2: Forward/Backward Fill**
- **Pros**: Preserves temporal patterns, works well for time series data, no data loss
- **Cons**: May not be realistic (carries forward outdated values), can create artificial patterns
- **Best for**: Time series data where values change gradually

**Method 3: Statistical Imputation (Mean/Median/Mode)**
- **Pros**: Simple, preserves overall distribution mean, works for any data type
- **Cons**: Reduces variance, ignores relationships between variables, may not reflect true patterns
- **Best for**: Large datasets where individual missing values won't significantly impact analysis

**Method 4: Interpolation**
- **Pros**: Creates realistic values based on trend, preserves patterns, smooth transitions
- **Cons**: May create false precision, assumes smooth relationships, computationally more complex
- **Best for**: Time series or ordered data with clear trends

**Key Considerations:**
1. **Amount of missing data**: High missingness may require dropping variables/observations
2. **Pattern of missingness**: Random vs systematic missing data requires different approaches
3. **Type of analysis**: Some methods better preserve statistical properties needed for specific analyses
4. **Domain knowledge**: Understanding why data is missing helps choose appropriate method
