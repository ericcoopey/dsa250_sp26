```python
# this is the format for loading a library and giving it an alias
import numpy as np

# if you create an alias, you must use it when you call the library
followees = np.array(['bob','mary','frank'])
balances = np.array([1526.73, 98774.91, 1000000, 4963.12, 123.19])

print(followees)
print(balances)
print(followees.dtype)
print(balances.dtype)
```
